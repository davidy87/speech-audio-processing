```python
import math
import numpy as np
import scipy
import scipy.io.wavfile
import scipy.signal
import matplotlib.pyplot as plt
```

```python
# Loading the audio file
signal = scipy.io.wavfile.read("signal.wav")
print(signal)
x1 = signal[1]
```

```python
# Mean subtraction
x0 = np.mean(x1)
x2 = x1 - x0
```

```python
# Pre-emphasis
x2_temp = np.append(0.0, x2[:len(x2) - 1])
x3 = x2 - 0.97 * x2_temp
```

```python
# Computing frames
def compute_frames(x3, L, S):
    N = math.ceil((len(x3) - L) / S)
    frames = np.zeros((N + 1, L))
    
    for k in range(N + 1):
        for n in range(L):
            if k * S + n < len(x3):
                frames[k][n] = x3[k * S + n]
    
    
    return frames
```

```python
# Applying the window function
def apply_window(frame):
    w = scipy.signal.hamming(len(frame))
    
    return frame * w
```

```python
# Computing the Fourier transform
def dft(windowed_frame):
    return scipy.fft.fft(windowed_frame, n=512)
```

```python
# Computing the magnitude and power spectra
def power_spectra(dft_frame):
    return np.square(np.abs(dft_frame))
```

```python
frames = compute_frames(x3, 400, 160)
X_pow = np.empty((frames.shape[0], 512))

for m in range(X_pow.shape[0]):
    X_pow[m] = power_spectra(dft(apply_window(frames[m])))
```

```python
# Mel-filterbank application
mel_filters = np.load("mel_filters.npy")
print(mel_filters.shape)

plt.plot(mel_filters.T)
plt.show()

def mel_filter(X_pow):
    return np.matmul(mel_filters, X_pow[:, :mel_filters.shape[1]].transpose()).transpose()

X_mel = mel_filter(X_pow)
print(X_mel.shape)
```

```python
# Taking the log
def logmel(X_mel):
    X_logmel = np.empty(X_mel.shape)
    
    for m in range(X_logmel.shape[0]):
        for k in range(X_logmel.shape[1]):
            X_logmel[m][k] = max(-50, np.log(X_mel[m, k]))
    
    return X_logmel

X_logmel = logmel(X_mel)
print(X_logmel.shape)
```

```python
# Computing the DCT and liftering
def dct(X_logmel):
    C = np.empty((X_logmel.shape[0], 13))
    
    for i in range(C.shape[0]):
        for k in range(X_logmel.shape[1]):
            C[i] = np.sum(X_logmel[k]) * math.cos(((math.pi * i) / X_logmel.shape[1]) * (k + 0.5))
            
    return C

print(dct(X_logmel).shape)
```

```python
"""
Putting it all together
"""
# plt.imshow(dct(X_logmel).transpose()[1:], origin="lower", aspect="auto")

print(dct(X_logmel))
mfcc_ref = np.load("reference_mfcc.npy")
print(mfcc_ref)
plt.imshow(mfcc_ref[:, 1:].transpose(), cmap="gray", origin="lower", aspect="auto")
```
